* **character** -> Cadena de texto.
* **numeric** -> Número continuo (Real).
* **integer** -> Número entero.
* **logical** -> Verdadero o falso.

## Ficheros en R 

Un archivo o fichero informático es un conjunto de bytes que son almacenados en un dispositivo. (Wikipedia)

### Archivo de excel

```r
library(readxl)
auto <- read_excel("Bases de datos/auto.xlsx", sheet = 2)
head(auto)
```

### Archivo de csv

```r
auto <- read.csv('Bases de datos/auto.csv')
head(auto)
```

### Archivo de txt

```r
auto <- read.delim('Bases de datos/auto.txt')
head(auto)
```

```r
auto.corregir <- read.delim('Bases de datos/auto_corregir.txt', skip = 2, sep = "/", header = F,
                            na.strings = c('-', '&', ' ', 'NA'))
head(auto.corregir)
```

# Estructuras de datos

### Vectores

Los vectores son arreglos ordenados en los cuales se puede almacenar información numérica (Cuantitativa), alfanumérico (Cualitativa), booleano (True, false), o combinaciones de los anteriores. En R pueden ser creados por medio de la instrucción `c()`que significa concatenar o por medio de las funciones `rep()`, `seq()` y `:`que permiten la creación de patrones.

```r
names(auto.corregir) <- c('No', 'mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                          'acceleration', 'model_year', 'car_name')
```

```r
head(auto.corregir)
```

```r
v1 <- 1:10 
v2 <- c(1, 3, 4)
v3 <- seq(from = 1, to = 7, by = 2)
v4 <- rep(4, 6)
```

### Data frames

Los data frame son estructuras de datos que permiten agrupar vectores diferentes tipos de datos (numérica, categórica, booleana) en un mismo objeto, pero que debe cumplir la condición de que todos los vectores deben tener la misma longitud. En R los data frame son generados por medio de la función `data.frame()`.

```r
class(auto)
```

# Recodificar datos

```r
auto$cylinders <- as.factor(auto$cylinders)
auto$car_name <- as.character(auto$car_name)
auto$model_year <- as.factor(auto$model_year)
```

```r
head(auto)
```

```r
print(paste('1', 'a'))
```

```r
print(paste(1:10, 'a'))
```

```r
levels(auto$cylinders)
```

```r
auto$cylinders <- factor(auto$cylinders,levels(auto$cylinders), labels = paste(levels(auto$cylinders), ' cyliders') )
```

```r
head(auto)
```

$$\frac{LimiteSuperior - LimiteInferior}{2}$$

```r
range(auto$weight)
```

```r
punto.corte <- (range(auto$weight)[2] +  range(auto$weight)[1]) / 2 
```

```r
auto$weight <= punto.corte
```

```r
auto$peso.categoria <- auto$weight <= punto.corte
```

```r
head(auto)
```

```r
auto$peso.categoria <- factor(auto$peso.categoria, c(T, F), c('Bajo', 'Alto'))
```

```r
head(auto)
```

# Subconjuntos de datos

```r
auto.4cilindros <- auto[auto$cylinders == '4  cyliders', ]
head(auto.4cilindros)
```

### Operadores 

* **x | y** <- Operador O

* **x & y** <- OPerador Y

* **!x** <- Negación

* **isTRUE(x)** <- X es verdad?

* **==** <- Comparación

* **<, < =, >, > =** <- Mayor, Mayor o igual, Menor y Menor o igual.

* **%in%** <- Permite determinar si uno o varios valores están contenidos en un vector.
